# EDA

### Correlation between the features value

```python
corr=df.corr()
corr
```

#### Plotting Heat Map for correlated value

```python
plt.figure(figsize=(10,5))
sns.heatmap(data=corr,annot=True,cmap="RdBu")
plt.title("correlation between the Data")
plt.show()
```

```python
num_columns = len(df.columns)
num_rows = (num_columns // 3) + (num_columns % 3 > 0)  

plt.figure(figsize=(10, 8))

for element in range(num_columns):
    plt.subplot(num_rows, 3, element + 1) 
    sns.histplot(df[df.columns[element]], kde=True)
    plt.title(f'Distribution for {df.columns[element]}')

plt.tight_layout()
plt.show()
```

#### mean of wine density based on wine pH & quality

```python
df1=df.groupby('density')[['pH','quality']].mean()
df1.head(20)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=df1.index,y=df1['pH'],color='b')
plt.title('Density vs Mean pH')
plt.xlabel('Density')
plt.ylabel('Mean pH')
plt.show()
```

```python
sns.scatterplot(x=df1.index, y=df1['quality'], color='r')
plt.title('Density vs Mean Quality')
plt.xlabel('Density')
plt.ylabel('Mean Quality')
plt.show()
```

```python
df2=df.groupby('fixed acidity')[['volatile acidity','citric acid']].max()
df2.head(20)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=df2.index,y=df2['volatile acidity'],linestyle=':', marker='o',color='Red')
plt.title("fixed acidity vs Max Volatile acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('Volatile Acidity')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=df2.index,y=df2['citric acid'],linestyle='-', marker='*',markersize=10,color='Green')
plt.title("fixed acidity vs Max Volatile acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('citric acid')
plt.show()
```

#### Sum the values for 'free sulfur dioxide' and 'total sulfur dioxide'

```python
free_sulfur_dioxide_sum = df['free sulfur dioxide'].sum()
total_sulfur_dioxide_sum = df['total sulfur dioxide'].sum()

sulfur_data = [free_sulfur_dioxide_sum, total_sulfur_dioxide_sum]
labels = ['free sulfur dioxide', 'Total Sulfur Dioxide']

plt.figure(figsize=(8, 6))
plt.pie(sulfur_data, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen'])
plt.title('Proportion of Free vs Total Sulfur Dioxide')
plt.axis('equal') 
plt.show()
```
